### Аналіз California Housing Dataset

California Housing Dataset містить дані про житлові райони Каліфорнії. Кожен рядок описує один район через характеристики населення, будинків і розташування.

#### Практична задача

За характеристиками району передбачити медіанну вартість житла.

#### Dataset

`../data/california_housing.csv`

#### Опис колонок

| Колонка | Що означає | Тип даних | Чи корисна для ML |
|---|---|---|---|
| `MedInc` | медіанний дохід у районі | число | так, це вхідна ознака |
| `HouseAge` | медіанний вік будинків у районі | число | так, це вхідна ознака |
| `AveRooms` | середня кількість кімнат на домогосподарство | число | так, це вхідна ознака |
| `AveBedrms` | середня кількість спалень на домогосподарство | число | так, це вхідна ознака |
| `Population` | населення району | число | так, це вхідна ознака |
| `AveOccup` | середня кількість мешканців на домогосподарство | число | так, це вхідна ознака |
| `Latitude` | географічна широта | число | так, може враховувати розташування |
| `Longitude` | географічна довгота | число | так, може враховувати розташування |
| `MedHouseVal` | медіанна вартість житла | число | так, але як цільова ознака |
