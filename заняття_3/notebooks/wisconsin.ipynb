### Аналіз Breast Cancer Wisconsin Dataset

Breast Cancer Wisconsin Dataset містить дані про характеристики клітинних ядер, отримані з цифрових зображень біопсії. Кожен рядок описує один зразок через числові ознаки форми, розміру та текстури клітин.

#### Практична задача

За характеристиками зразка передбачити діагноз: доброякісна пухлина (`B`) або злоякісна пухлина (`M`).

#### Dataset

`../../data/breast_cancer_wisconsin.csv`

#### Опис колонок

| Колонка | Що означає | Тип даних | Чи корисна для ML |
|---|---|---|---|
| `id` | ідентифікатор зразка | число | ні, це службова колонка |
| `diagnosis` | діагноз: `B` - доброякісна, `M` - злоякісна | категорія | так, але як цільова ознака |
| `radius_mean` | середній радіус клітинного ядра | число | так, це вхідна ознака |
| `texture_mean` | середня текстура клітинного ядра | число | так, це вхідна ознака |
| `perimeter_mean` | середній периметр клітинного ядра | число | так, це вхідна ознака |
| `area_mean` | середня площа клітинного ядра | число | так, це вхідна ознака |
| `smoothness_mean` | середня гладкість контуру | число | так, це вхідна ознака |
| `compactness_mean` | середня компактність | число | так, це вхідна ознака |
| `concavity_mean` | середня увігнутість контуру | число | так, це вхідна ознака |
| `concave points_mean` | середня кількість увігнутих точок | число | так, це вхідна ознака |
| `symmetry_mean` | середня симетрія | число | так, це вхідна ознака |
| `fractal_dimension_mean` | середня фрактальна розмірність | число | так, це вхідна ознака |
| `*_se` | стандартна похибка відповідної ознаки | число | так, це вхідні ознаки |
| `*_worst` | найбільше або найгірше значення відповідної ознаки | число | так, це вхідні ознаки |
| `Unnamed: 32` | порожня колонка з CSV | число/порожня | ні, її варто видалити |
