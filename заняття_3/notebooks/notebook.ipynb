```python
print("Кількість значень:", df['amount'].count())
print("Середнє:", df['amount'].mean())
print("Медіана:", df['amount'].median())
print("Мінімум:", df['amount'].min())
print("Максимум:", df['amount'].max())
```

```python
# Завдання 6
mean_amount = df['amount'].mean()
median_amount = df['amount'].median()
```
