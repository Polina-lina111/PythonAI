### Аналіз Diabetes Dataset

Diabetes Dataset зі scikit-learn містить медичні дані про пацієнтів із діабетом. Кожен рядок описує одного пацієнта через 10 числових ознак: вік, стать, індекс маси тіла, середній артеріальний тиск і шість показників аналізу крові.

#### Практична задача

За медичними характеристиками пацієнта передбачити кількісний показник прогресування діабету через один рік після початкового вимірювання.

#### Dataset

`../../data/diabetes.csv`

#### Опис колонок

| Колонка | Що означає | Тип даних | Чи корисна для ML |
|---|---|---|---|
| `age` | вік пацієнта, стандартизоване значення | число | так, це вхідна ознака |
| `sex` | стать пацієнта, стандартизоване значення | число | так, це вхідна ознака |
| `bmi` | індекс маси тіла, стандартизоване значення | число | так, це вхідна ознака |
| `bp` | середній артеріальний тиск, стандартизоване значення | число | так, це вхідна ознака |
| `s1` | перший показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `s2` | другий показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `s3` | третій показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `s4` | четвертий показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `s5` | пʼятий показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `s6` | шостий показник аналізу крові, повʼязаний із сироватковими вимірюваннями | число | так, це вхідна ознака |
| `target` | кількісний показник прогресування діабету через один рік | число | так, але як цільова ознака |
