```python
import nltk

nltk.download("movie_reviews")
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
from nltk.corpus import movie_reviews
from nltk.tokenize import word_tokenize
import pandas as pd
```

Отримання списку відгуків

```python
pos_ids = movie_reviews.fileids("pos")[:10]
neg_ids = movie_reviews.fileids("neg")[:10]

selected_ids = pos_ids + neg_ids
```

Токенізація відгуків

```python
rows = []

for fileid in selected_ids:
    text = movie_reviews.raw(fileid)
    tokens = word_tokenize(text)
    category = fileid.split("/")[0]

    rows.append({
        "fileid": fileid,
        "category": category,
        "token_count": len(tokens),
        "first_20_tokens": tokens[:20],
        "tokens": tokens
    })

df = pd.DataFrame(rows)

df[["fileid", "category", "token_count", "first_20_tokens"]]
```

Якщо first_20_tokens важко читати в таблиці

```python
df["first_20_tokens_text"] = df["first_20_tokens"].apply(lambda tokens: " ".join(tokens))

df[["fileid", "category", "token_count", "first_20_tokens_text"]]
```

Базова статистика

```python
total_tokens = df["token_count"].sum()
avg_tokens = df["token_count"].mean()
min_tokens = df["token_count"].min()
max_tokens = df["token_count"].max()

print("Загальна кількість токенів:", total_tokens)
print("Середня кількість токенів:", avg_tokens)
print("Мінімальна довжина:", min_tokens)
print("Максимальна довжина:", max_tokens)
```

Середня довжина позитивних і негативних відгуків

```python
df.groupby("category")["token_count"].mean()
```

Зібрати всі токени

```python
all_tokens = [token for tokens in df["tokens"] for token in tokens]
```

 перші 30 токенів, які не складаються тільки з літер:

```python
strange_tokens = [token for token in all_tokens if not token.isalpha()]

strange_tokens[:30]
```

які такі токени трапляються найчастіше:

```python
from collections import Counter

Counter(strange_tokens).most_common(20)
```
