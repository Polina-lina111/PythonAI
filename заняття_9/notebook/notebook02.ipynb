```python
import torch
from torch import nn
from torch import optim
import pandas as pd
import numpy as np
```

```python
X = torch.randn(100, 1)
display(X.shape)
y = 3.0 * X + 2.0 + 0.2 * torch.randn(100, 1)
display(y.shape)
```

```python
model = nn.Linear(1, 1)
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
optimizer = optim.Adam(model.parameters(), lr=0.1)

history = []

for epoch in range(100):
    y_pred = model(X)

    loss = loss_fn(y_pred, y)

    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

    history.append({
        "epoch": epoch,
        "loss": loss.item(),
        "weight": model.weight.item(),
        "bias": model.bias.item()
    })

history = pd.DataFrame(history)
history.head()
```
