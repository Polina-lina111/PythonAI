```python
import pandas as pd

from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
messages = [
    "My order is two weeks late and I still have not received it.",
    "I cannot log into my account.",
    "My credit card was charged twice.",
    "Thank you for the fast delivery!",
    "Can you tell me when my order will arrive?",
    "The product arrived damaged.",
    "Everything works perfectly. Thank you!",
    "I forgot my password and cannot sign in.",
    "Where can I change my shipping address?",
    "I am very disappointed with your service."
]
```

```python
results = classifier(messages)
```

```python
rows = []

for i in range(len(messages)):

    if results[i]["label"] == "NEGATIVE" and results[i]["score"] > 0.90:
        priority = "high"
    else:
        priority = "normal"

    rows.append({
        "message": messages[i],
        "label": results[i]["label"],
        "score": round(results[i]["score"], 3),
        "priority": priority
    })

df = pd.DataFrame(rows)

df
```
