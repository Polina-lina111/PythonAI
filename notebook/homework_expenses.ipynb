```python
expenses = [1200, 1500, 800, 600, 1000, 400, 900]
expenses
```

```python
total = sum(expenses)
average = sum(expenses) / len(expenses)
maximum = max(expenses)
minimum = min(expenses)

days_above_1000 = len([x for x in expenses if x > 1000])

total, average, maximum, minimum, days_above_1000
```

```python
print("Показник | Значення")
print("----------------------")
print("Загальна сума:", total)
print("Середнє значення:", average)
print("Максимальна витрата:", maximum)
print("Мінімальна витрата:", minimum)
print("Днів > 1000:", days_above_1000)
```
