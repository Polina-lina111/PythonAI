```python
from transformers import pipeline
import pandas as pd
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
messages = [
    "Thank you for your help!",
    "Could you please explain this task again?",
    "The weather is nice today.",
    "I finished my homework.",
    "The product works as expected.",
    "I am very disappointed with your service.",
    "This is the worst application ever.",
    "Your support is useless.",
    "I hate this product.",
    "Everything is terrible."
]
```

```python
results = classifier(messages)
```

```python
df = pd.DataFrame({
    "message": messages,
    "label": [r["label"] for r in results],
    "score": [r["score"] for r in results]
})
```

```python
df["needs_review"] = (
    (df["label"] == "NEGATIVE") &
    (df["score"] > 0.90)
)
```

Висновок

Модель sentiment-analysis може швидко знаходити повідомлення з негативним емоційним забарвленням і позначати їх для перевірки. Це корисно як перший автоматичний фільтр.

Однак такий підхід не можна використовувати як єдине рішення, оскільки модель може неправильно визначити сарказм, жарти, контекст або ввічливу критику. Тому остаточне рішення повинна приймати людина або більш складна система модерації.
