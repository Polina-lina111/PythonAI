```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Ім'я": ["Аня", "Богдан", "Іван", "Оля", "Марія", "Петро", "Софія", "Данило"],
    "Проект": ["AI", "Web", "AI", "ML", "Web", "ML", "AI", "Web"],
    "Години": [10, 15, 8, 20, 12, 7, 14, 9]
})

df
```

```python
df.groupby("Ім'я")["Години"].sum()
```

```python
df.groupby("Проект")["Години"].sum()
```

```python
hours_by_employee = df.groupby("Ім'я")["Години"].sum()

print("Співробітник:", hours_by_employee.idxmax())
print("Години:", hours_by_employee.max())
```
