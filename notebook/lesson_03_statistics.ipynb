```python
import pandas as pd

houses_df = pd.read_csv('../data/california_housing.csv')

houses_df.head()
```

```python
selected_columns = ['MedHouseVal', 'MedInc', 'HouseAge', 'AveRooms', 'Population']

statistics_df = pd.DataFrame({
    'mean': houses_df[selected_columns].mean(),
    'median': houses_df[selected_columns].median(),
    'mode': houses_df[selected_columns].mode().iloc[0],
    'variance': houses_df[selected_columns].var(),
    'std': houses_df[selected_columns].std()
})

statistics_df.round(3)
```

```python
feature_pairs = [
    ('MedInc', 'MedHouseVal'),
    ('AveRooms', 'MedHouseVal'),
    ('HouseAge', 'MedHouseVal')
]

correlations_df = pd.DataFrame(
    [
        {
            'feature_1': feature_1,
            'feature_2': feature_2,
            'correlation': houses_df[feature_1].corr(houses_df[feature_2])
        }
        for feature_1, feature_2 in feature_pairs
    ]
)

correlations_df.round(3)
```

```python
## Інтерпретація кореляцій

# MedInc і MedHouseVal мають позитивну кореляцію. Це означає, що райони з вищим доходом населення зазвичай мають дорожче житло.

# AveRooms і MedHouseVal також мають позитивний зв'язок. Більша кількість кімнат може свідчити про більші або дорожчі будинки.

# HouseAge і MedHouseVal мають слабший зв'язок. Вік будинків сам по собі не завжди визначає їхню вартість.
```

```python
input_columns = [
    'MedInc',
    'HouseAge',
    'AveRooms',
    'Population'
]

mean_median_difference_df = pd.DataFrame({
    'mean': houses_df[input_columns].mean(),
    'median': houses_df[input_columns].median()
})

mean_median_difference_df['absolute_difference'] = (
    mean_median_difference_df['mean'] -
    mean_median_difference_df['median']
).abs()

mean_median_difference_df.sort_values(
    'absolute_difference',
    ascending=False
).round(3)
```

```python
## Різниця між середнім і медіаною

# Для колонки Population середнє значення помітно більше за медіану.

# Це може означати, що більшість районів мають відносно невелике населення, але є окремі райони з дуже великою кількістю мешканців.

# Такі великі значення підвищують середнє, але майже не впливають на медіану.
```
